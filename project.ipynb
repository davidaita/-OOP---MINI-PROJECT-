```python
# Parent Class
class SmartDevice:
    def __init__(self, device_id, power_status, name):
        self.__device_id = device_id
        self.__power_status = power_status
        self.name = name
    
    # Method to turn device on
    def turn_on(self):
        self.__power_status = 'On'
        print(f'{self.name} is ON')
    
    # Method to turn device off
    def turn_off(self):
        self.__power_status = 'Off'
        print(f'{self.name} is OFF')
    
    # Method to display device information
    def display_info(self):
        print(f'{self.name} Info')
        print(f'Device ID: {self.__device_id} \nPower Status: {self.__power_status} \nDevice Name: {self.name} \n')
    
    # getter to display device id
    @property
    def device_id(self):
        print(f'Device ID: {self.__device_id}')
        
    # Setter to set device id
    @device_id.setter
    def device_id(self,value):
        self.__device_id = value
    
    # getter to display device power status
    @property
    def power_status(self):
        print(f'Device Power Status: {self.__power_status}')
    
    # setter to set device display power status
    @power_status.setter
    def power_status(self, value):
        self.__power_status = value
    
        
```

```python
# Child Class
class TemperatureSensor(SmartDevice):
    def __init__(self,device_id, power_status, name, temperature):
        super().__init__(device_id, power_status, name)
        self.temperature = temperature
    
    # Method to read device temperature
    def read_temperature(self):
        print(f'The temperature is {self.temperature} degrees')
```

```python
# Child Class
class SecurityCamera(SmartDevice):
    def __init__(self,device_id, power_status, name, recording_status):
        super().__init__(device_id, power_status, name)
        self.recording_status = recording_status
    
    # Method to start device recording
    def start_recording(self):
        self.recording_status = True
        print(f'{self.name} has STARTED recording')
    
    # Method to stop device recording
    def stop_recording(self):
        self.recording_status = False
        print(f'{self.name} has STOPPED recording')
    
```

```python
# Child Class
class SmartLight(SmartDevice):
    def __init__(self,device_id, power_status, name, brightness):
        super().__init__(device_id, power_status, name)
        self.brightness = brightness
    
    # Method to increase brightness
    def increase_brightness(self):
        self.brightness +=1
        print(f'Brightness level: {self.brightness}')
    
    # Method to decrease brightness
    def decrease_brightness(self):
        self.brightness -=1
        print(f'Brightness level: {self.brightness}')
```

```python
# Devices setup
smartFridge = SmartDevice('ID213', 'On', 'Hisense Fridge')
tempDevice = TemperatureSensor('ID213', 'On', 'Hisense Temperature Device', 300)
securityCamera = SecurityCamera('ID4536', 'On', 'Hisense CCTV', True)
smartLight = SmartLight('ID453', 'On', 'Hisense Disco Light',5  )


# user menu
while True:
    print('')
    print('1. Display Device Information')
    print('2. Turn on Device')
    print('3. Turn Device Off')
    print('4. Read Temperature')
    print('5. Adjust Brightness')
    print('6. Start Recording')
    print('7. Exit')
    print('')
    
    # menu response
    response = int(input('Enter your response: '))
    
    if response == 1:
        smartFridge.display_info()
        tempDevice.display_info()
        securityCamera.display_info()
        smartLight.display_info()
    elif response == 2:
        smartFridge.turn_on()
        tempDevice.turn_on()
        securityCamera.turn_on()
        smartLight.turn_on()
    elif response == 3:
        smartFridge.turn_off()
        tempDevice.turn_off()
        securityCamera.turn_off()
        smartLight.turn_off()
    elif response == 4:
        tempDevice.read_temperature()
    elif response == 5:
        res = int(input('Enter 1 to increase, 2 to decrease: '))
        if res == 1:
            smartLight.increase_brightness()
        elif res == 2:
            smartLight.decrease_brightness()
        else:
            print('Invalid input')
    elif response == 6:
        securityCamera.start_recording()
    elif response == 7:    
        break
    else:
        print('Input valid response!')

```
